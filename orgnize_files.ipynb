```python
from pathlib import Path
import shutil

fake_root = Path('datasets/HOI4D/')

src_dirs = [
    "HOI4D_release",
    "HOI4D_depth_video",
    "HOI4D_annotations",
]

base_root = fake_root / src_dirs[0]

for zy_dir in base_root.glob("ZY2021080000*"):
    for t_dir in zy_dir.glob("H*/C*/N*/S*/s*/T*"):

        rel_path = t_dir.relative_to(base_root)

        dst_t_dir = fake_root / rel_path

        dst_t_dir.mkdir(parents=True, exist_ok=True)

        for src_name in src_dirs:
            src_base = fake_root / src_name
            src_t_dir = src_base / rel_path

            if not src_t_dir.exists():
                continue

            for item in src_t_dir.iterdir():
                src_item = item
                dst_item = dst_t_dir / item.name

                if dst_item.exists():
                    print("SKIP (exists):", dst_item)
                    continue

                print("MOVE:", src_item, "->", dst_item)
                shutil.move(str(src_item), str(dst_item))
```
